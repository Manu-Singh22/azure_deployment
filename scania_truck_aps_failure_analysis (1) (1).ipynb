```python
df.replace(to_replace='na',value=np.nan,inplace=True)
#so we have define a hyperparameter i.e. sparse_threshold.
#drop nan values or fill it

dropped_df=df.dropna(axis=1,thresh=31000,inplace=False)
```

```python
dropped_df
```

```python
#so we can say that it is highly unbalanced dataset.
#why unbalanced data is serious problem for our clf algos lets take naive bayes
#we will get p(class=pos)=1000/60000 that is a very small value so this prior will vanish the overall probability of P(class=pos|X).
#so our algo will mostly predict class=neg and that will cause misclassification and now we have to deal with type1 and type2 error.
#so we have to balance our data using SMOTE.
#first we have to preprocess the data.
#checking sparse and dense columns.
#checking the number of unique values




unique_vals_dict=dict()



for col in dropped_df.columns[1:]:
    
    unique_vals_dict[col]= dropped_df[col].unique().shape[0]
```

```python
unique_vals_dict
```

```python
sparse_threshold=0.01
sparse_columns=list()


for single_column in unique_vals_dict.keys():
    
    if unique_vals_dict[single_column]<= sparse_threshold*dropped_df.shape[0]:
        sparse_columns.append(single_column)
```

```python
len(sparse_columns)
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
for single_column in sparse_columns:
    col=np.array(dropped_df[single_column]).reshape(dropped_df[single_column].shape[0],1)
    imputer.fit(col)
    dropped_df[single_column]=imputer.transform(col)
```

```python
dropped_df
```

```python
dense_columns=list()
for col in dropped_df.columns[1:]:
    if col not in sparse_columns:
        dense_columns.append(col)
        
```

```python
len(dense_columns)
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
for single_column in dense_columns:
    col=np.array(dropped_df[single_column]).reshape(dropped_df[single_column].shape[0],1)
    imputer.fit(col)
    dropped_df[single_column]=imputer.transform(col)
```

```python
#here we have completed our feature selection
#now we can apply SMOTE for synthetic sample point for minority class.
#SMOTE synthetic minority oversampling technique, in this technique we create synthetic sample point for minority class using
#the existing minority class. It baically use the KNN algo for oversampling
#one downside of the this approach is, it don't consider majority class while performing the KNN.
#it is not prefered to generate the balance dataset by oversampling the minority class sample points equal to majority class.
#it is prefered oversample the minority class and undersample the majority class.
#we will be using the the pipeline for the transformation of dataset.
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
k_vals_smote=[1,2,3,4,5,6,7]
for k in k_vals_smote:
    #defining the pipeline
    #model1=DecisionTreeClassifier()
    model2=GaussianNB()
    
    over=SMOTE(sampling_strategy=0.1,k_neighbors=k,random_state=1)
    under=RandomUnderSampler(sampling_strategy=0.5)
    steps=[('over',over),('under',under),('model2',model2)]
    pipeline=Pipeline(steps=steps)
    
    
    
    # evaluate the pipeline
    cv = RepeatedStratifiedKFold(n_splits=10,n_repeats=3,random_state=1)
    scores= cross_val_score(pipeline,X,y,scoring='roc_auc',cv=cv,n_jobs=-1)
    print(scores)
    s=scores.reshape(30,1)
    
    score =np.mean(s,axis=0)
    
    print('k={0}, mean of roc_auc: {1}'.format(k,score))
    
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RepeatedStratifiedKFold
def best_hyperparameter_models(X,y):
    algos={
        'Adaboost':{
            'model':AdaBoostClassifier(),
            'param_grid':{'n_estimators':[60,70,100],
                          'algorithm':['SAMME','SAMME.R']
                           }
        },
        'SVC':{
            'model':SVC(),
            'param_grid':{'C':[25.0,30.0,100.0],
                          'kernel':['linear','rbf'],
                          'gamma':['scale','auto']
                }
        },
        "Randomforest":{ 
            'model':RandomForestClassifier(),
            'param_grid':{'n_estimators':[100,80,50,30],
                          'criterion':['gini','entropy'],
                          'max_features':['auto','sqrt','log2']
            }
                        }
    }
    
    for algo_name,config in algos.items():
        cv=RepeatedStratifiedKFold(n_splits=10,n_repeats=3,random_state=1)
        clf=RandomizedSearchCV(config['model'],config['param_grid'],cv=cv,scoring='roc_auc',n_jobs=-1,return_train_score=True)
        clf.fit(X,y)
        print('Results for best estimators:{}',clf.cv_results_)
        
        return (clf.best_estimator_,clf.best_params_,clf.best_score_)    
```
